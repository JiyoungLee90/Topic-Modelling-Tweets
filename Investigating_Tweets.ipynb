```python
import pandas as pd
import numpy as np
```

```python
tweet_df = pd.read_csv('tweet_climate_change_processed.csv')
tweet_df
```

```python
tweet_df['Processed_token'] = tweet_df['Processed_token'].apply(eval)
tweet_df['Bigram'] = tweet_df['Bigram'].apply(eval)
```

```python
def find_tweets(token):
    idx_list = [idx for idx, tweet in enumerate(tweet_df['Bigram']) if token in tweet]
    return list(tweet_df['Tweet'].iloc[idx_list])
```

## Retweet count

### All users

```python
tweet_df.duplicated(subset='Tweet', keep='first').sum()
```

```python
counts = tweet_df.pivot_table(index=['Tweet'], aggfunc='size')
counts = pd.DataFrame(counts) # Convert Series to DataFrame
counts.index.name = 'tweet'
counts.reset_index(inplace=True) # Change row names to be a column
counts.columns = ['tweet', 'counts']
count_df = counts.merge(counts) # Merge dataframes on common column
```

```python
count_df['percent'] = count_df.counts/len(tweet_df)*100
```

```python
count_df.loc[21314, 'counts'] = count_df.loc[21314, 'counts'] + 149
```

```python
count_df = count_df.drop(21316)
```

```python
count_df.sort_values(by = ['counts'], ascending = False)[0:10]
```
